```python
import random
import time

def insertion_sort (cards):
    for j in range (1,len(cards)):
        key=cards[j]
        i=j-1
        while i>=0 and cards[i]>key:
            cards[i+1]=cards[i]
            i=i-1
        cards[i+1]=key
    return cards
timelist=0
t=100
for i in range(0,t):
    n=100
    state=list(range(1,n))
    random.shuffle(state)

    starttime=time.time()
    insertion_sort(state)
    timeconsumption=time.time()-starttime
    timelist=timelist+timeconsumption
    print("operation took" + "{0:.20f}".format(timeconsumption)+ "seconds")
    print(state)
print(timelist/t)
```
